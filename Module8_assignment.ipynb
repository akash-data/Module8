Q4: Explain the use of the groupby __( )__ method in Pandas?

Ans->

The `groupby()` method in Pandas is a fundamental tool for data analysis, enabling the "split-apply-combine" strategy. This strategy involves:

**Applying:**

  * __Aggregation:__ Computing a summary statistic for each group (e.g., `sum()`, `mean()`, `count()`, `min()`, `max()`).
  * __Transformation:__ Applying a function to each group and returning a result with the same index as the original group (e.g., standardizing data within groups, filling missing values based on group statistics).
  * __Filtration:__ Discarding entire groups based on a condition evaluated on the group (e.g., removing groups with fewer than a certain number of members).

**Combining:**

Merging the results from the applied operations on each group back into a single Series or DataFrame.



__Example__




```python
import pandas as pd

# Sample dataset with more features
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'C'],
    'Subcategory': ['X', 'X', 'Y', 'Y', 'Z', 'X', 'Z', 'Y', 'X'],
    'Value1': [10, 15, 12, 8, 20, 11, 5, 18, 7],
    'Value2': [100, 200, 150, 120, 180, 90, 50, 210, 80]
}
df = pd.DataFrame(data)

# Group by both 'Category' and 'Subcategory'
grouped = df.groupby(['Category', 'Subcategory'])

# Apply multiple aggregation functions
aggregated = grouped.agg({
    'Value1': ['mean', 'sum', 'count'],
    'Value2': lambda x: x.max() - x.min()  # Custom function: range
})

# Optional: Rename the custom lambda column
aggregated.columns = ['Value1_mean', 'Value1_sum', 'Value1_count', 'Value2_range']

# Sort the result by 'Value1_sum' descending
aggregated_sorted = aggregated.sort_values(by='Value1_sum', ascending=False)

print(aggregated_sorted)
```

```python
# Q2: Write a Pandas script to find the mean of a specific column in a DataFrame?

import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values and multiple columns
data = {
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'Salary': [70000, 50000, np.nan, 65000, 48000, 72000, np.nan],
    'Experience': [3, 5, 2, 4, 6, 1, 3]
}

df = pd.DataFrame(data)

# Step 1: Drop rows where Salary is NaN
df_clean = df.dropna(subset=['Salary'])

# Step 2: Filter employees with more than 3 years of experience
filtered_df = df_clean[df_clean['Experience'] > 3]

# Step 3: Group by Department and calculate mean Salary
mean_salaries = filtered_df.groupby('Department')['Salary'].mean()

print("Mean Salary by Department (with >3 years experience):\n", mean_salaries)
```

```python
# Q3: Create a scatter plot using Matplotlib?

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = 0.5 * x + np.random.normal(0, 10, n)  # y = 0.5x + noise

# Create categories for coloring
categories = np.random.choice(['Group A', 'Group B'], size=n)

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Group': categories})

# Plot
plt.figure(figsize=(10, 6))

for group, color, marker in zip(['Group A', 'Group B'], ['blue', 'green'], ['o', 's']):
    subset = df[df['Group'] == group]
    plt.scatter(subset['X'], subset['Y'],
                label=group,
                c=color,
                marker=marker,
                s=70,
                alpha=0.7,
                edgecolors='black')

# Linear regression fit
model = LinearRegression()
model.fit(df[['X']], df['Y'])
y_pred = model.predict(df[['X']])

# Plot regression line
plt.plot(df['X'], y_pred, color='red', linewidth=2, label='Trendline')

# Annotate a point
max_y_index = df['Y'].idxmax()
plt.annotate('Highest Y', xy=(df.loc[max_y_index, 'X'], df.loc[max_y_index, 'Y']),
             xytext=(df.loc[max_y_index, 'X'] + 5, df.loc[max_y_index, 'Y'] + 5),
             arrowprops=dict(arrowstyle='->', color='black'))

# Customizations
plt.title("Advanced Scatter Plot with Regression Line", fontsize=14)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Q4: How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a complex DataFrame
np.random.seed(0)
df = pd.DataFrame({
    'Math': np.random.randint(50, 100, 100),
    'Physics': np.random.randint(30, 100, 100),
    'Chemistry': np.random.randint(40, 100, 100),
    'English': np.random.randint(60, 100, 100),
    'Attendance (%)': np.random.uniform(70, 100, 100),
    'Participation': np.random.randint(0, 10, 100),
    'Grade': np.random.choice(['A', 'B', 'C'], size=100)
})

# 2. Encode categorical variable (optional step)
df['Grade_Code'] = df['Grade'].map({'A': 3, 'B': 2, 'C': 1})

# 3. Introduce some missing values
df.loc[5:10, 'Physics'] = np.nan
df.loc[20:25, 'Math'] = np.nan

# 4. Handle missing data (optional: drop or fill)
df_clean = df.dropna()  # or df.fillna(df.mean())

# 5. Compute correlation matrix
corr_matrix = df_clean.corr(numeric_only=True)

# 6. Create a mask for the upper triangle (for cleaner look)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# 7. Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# 8. Draw the heatmap
sns.heatmap(corr_matrix,
            mask=mask,
            annot=True,
            fmt=".2f",
            cmap="coolwarm",
            linewidths=0.5,
            cbar_kws={"shrink": 0.8},
            vmin=-1,
            vmax=1)

# 9. Customize plot
plt.title("Correlation Matrix Heatmap (Cleaned Data)", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# 10. Show plot
plt.show()
```

```python
# Q5: Generate a bar plot using Plotly?

import plotly.graph_objects as go

departments = ['IT', 'HR', 'Finance', 'Marketing', 'Sales']
q1 = [120000, 80000, 95000, 70000, 150000]
q2 = [130000, 85000, 98000, 75000, 160000]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=departments,
    y=q1,
    name='Q1 Revenue',
    marker_color='indigo',
    text=q1,
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=departments,
    y=q2,
    name='Q2 Revenue',
    marker_color='darkorange',
    text=q2,
    textposition='outside'
))

# Layout customization
fig.update_layout(
    barmode='group',
    title='Revenue Comparison by Department',
    xaxis_title='Department',
    yaxis_title='Revenue ($)',
    template='plotly_dark',
    legend=dict(x=0.8, y=1.1),
    margin=dict(t=50, l=50, r=50, b=50)
)

fig.show()
```

```python
# Q6: Create a DataFrame and add a new column based on an existing column?

import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['Akash', 'Siya', 'S1', 'Praveen', 'Prasant'],
    'Marks': [92, 76, 58, 85, 43]
}

df = pd.DataFrame(data)

# Step 2: Add a new column 'Grade' based on 'Marks'
def assign_grade(mark):
    if mark >= 90:
        return 'A+'
    elif mark >= 80:
        return 'A'
    elif mark >= 70:
        return 'B'
    elif mark >= 60:
        return 'C'
    elif mark >= 50:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Marks'].apply(assign_grade)

# Display the DataFrame
print(df)
```

```python
# Q10: Create a histogram using Seaborn to visualize a distribution?

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a synthetic dataset (e.g., student scores with outliers)
np.random.seed(42)
scores = np.concatenate([
    np.random.normal(70, 10, 200),  # normal group
    np.random.normal(90, 3, 10),    # high outliers
    np.random.normal(40, 3, 5)      # low outliers
])

df = pd.DataFrame({'Scores': scores})

# Create the histogram with Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df,
             x='Scores',
             bins=20,
             kde=True,
             color='skyblue',
             edgecolor='black',
             stat='density')  # use density for overlaying KDE

# Customize the plot
plt.title('Distribution of Student Scores', fontsize=16)
plt.xlabel('Score')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.5)

# Add a vertical line for the mean
mean_score = df['Scores'].mean()
plt.axvline(mean_score, color='red', linestyle='--', label=f'Mean: {mean_score:.2f}')

# Show legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
